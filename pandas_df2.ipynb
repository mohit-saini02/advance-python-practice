```python
## in dataframe: 

points = {
    '1st_point':[(3,4), (-6,5) , (0,0),(10,1) , (4,5)],
    '2nd_point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
}
points_df = pd.DataFrame(points)
points_df
```

```python
def euclidean(row):
    p1 = row['1st_point']
    p2  = row['2nd_point']
    return  ((p1[0] - p2[0])**2  + (p1[1] - p2[1])**2)**0.5

```

```python
points_df['distance'] = points_df.apply(euclidean , axis=1)
```
